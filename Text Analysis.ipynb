```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import os

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Function to clean text using stop words lists
def clean_text(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    cleaned_words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
    cleaned_text = ' '.join(cleaned_words)
    return cleaned_text

# Function to create positive and negative word dictionaries
def create_word_dict(file_path):
    word_dict = {}
    with open(file_path, 'r') as file:
        for word in file:
            word = word.strip().lower()
            word_dict[word] = True
    return word_dict

# Function to perform sentiment analysis and extract derived variables
def sentiment_analysis(text, positive_dict, negative_dict):
    tokens = word_tokenize(text)
    total_words = len(tokens)
    
    positive_score = sum(1 for word in tokens if word.lower() in positive_dict)
    negative_score = sum(1 for word in tokens if word.lower() in negative_dict)
    
    polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 0.000001)
    subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)
    
    return {
        'Positive Score': positive_score,
        'Negative Score': -negative_score,  # Multiplying by -1 to make it positive
        'Polarity Score': polarity_score,
        'Subjectivity Score': subjectivity_score
    }

# Load positive and negative word dictionaries
positive_dict = create_word_dict('Aayushi/positive-words.txt')  # Update with the correct path
negative_dict = create_word_dict('Aayushi/negative-words.txt')  # Update with the correct path

# Folder containing input text files
input_folder = 'Aayushi/'  # Update with the name of the folder containing input text files

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        filepath = os.path.join(input_folder, filename)
        
        # Read the input text from the file
        with open(filepath, 'r', encoding='latin1') as file:
            text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(text)
        
        # Perform sentiment analysis and extract derived variables
        derived_variables = sentiment_analysis(cleaned_text, positive_dict, negative_dict)
        
        # Print the derived variables for the current file
        print(f"Analysis for file: {filename}")
        for variable, value in derived_variables.items():
            print(f"{variable}: {value}")
        print()
```

```python

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
import os
from nltk.corpus import stopwords

# Function to calculate Gunning Fox index
def gunning_fox_index(text):
    # Tokenize text into sentences and words
    sentences = sent_tokenize(text)
    words = word_tokenize(text.lower())
    
    # Remove stopwords and lemmatize words
    stop_words = set(stopwords.words('english'))
    words = [WordNetLemmatizer().lemmatize(word) for word in words if word.isalnum() and word not in stop_words]
    
    # Calculate average sentence length
    avg_sentence_length = len(words) / len(sentences)
    
    # Identify complex words (words with more than 2 syllables)
    syllable_count = lambda word: sum([1 for char in word if char.lower() in 'aeiouy'])
    complex_words = [word for word in words if syllable_count(word) > 2]
    
    # Calculate percentage of complex words
    percentage_complex_words = len(complex_words) / len(words)
    
    # Calculate Fog Index
    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)
    
    return fog_index, percentage_complex_words, avg_sentence_length

# Folder containing input text files
input_folder = 'Aayushi/'

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        filepath = os.path.join(input_folder, filename)
        
        # Read the input text from the file
        with open(filepath, 'r', encoding='latin1') as file:
            text = file.read()
        
        # Calculate Gunning Fox index for the current file
        index, percentage_complex_words, avg_sentence_length = gunning_fox_index(text)
        print(f"Gunning Fox Index for {filename}: {index}")
        print(f"Percentage of complex words for {filename}: {percentage_complex_words}")
        print(f"Average sentence length for {filename} : {avg_sentence_length}")
```

```python
from nltk.corpus import cmudict
import os
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk


def count_syllables(word, d):
    """Count syllables in a word using the CMU Pronouncing Dictionary."""
    if word.lower() in d:
        return sum(len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()])
    else:
        return 0 

# Function to calculate Gunning Fox index
def gunning_fox_index(text):
    """Calculate Gunning Fox index for a given text."""
    # Tokenize text into sentences and words
    sentences = sent_tokenize(text)
    words = word_tokenize(text.lower())
    
    # Initialize CMU Pronouncing Dictionary
    d = cmudict.dict()
    
    # Calculate total number of words and sentences
    total_words = len(words)
    total_sentences = len(sentences)
    
    # Count the number of complex words (words with more than two syllables)
    num_complex_words = sum(1 for word in words if count_syllables(word, d) > 2)
                
    # Count the number of syllables in each word            
    num_syllables = sum(count_syllables(word, d) for word in words)
    
    # Calculate average number of words per sentence
    avg_words_per_sentence = total_words / total_sentences
    
    return avg_words_per_sentence, num_complex_words, num_syllables
   

# Folder containing input text files
input_folder = 'Aayushi/'

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        filepath = os.path.join(input_folder, filename)
        
        # Read the input text from the file
        try:
            with open(filepath, 'r', encoding='latin1') as file:
                text = file.read()
        except FileNotFoundError:
            print(f"File not found: {filename}")
            continue
        except Exception as e:
            print(f"Error reading file {filename}: {e}")
            continue
        
        # Calculate Gunning Fox index for the current file
        try:
            avg_words_per_sentence, num_complex_words, num_syllables = gunning_fox_index(text)
        except Exception as e:
            print(f"Error processing file {filename}: {e}")
            continue
        
        print(f"File: {filename}")
        print(f"Average Number of Words Per Sentence: {avg_words_per_sentence}")
        print(f"Number of Complex Words: {num_complex_words}")
        print(f"Number of Syllables: {num_syllables}\n")
```

```python
import re
import os

# Function to count personal pronouns in the text
def count_personal_pronouns(text):
    # Define regex pattern for personal pronouns
    pattern = r'\b(I|we|my|ours|us)\b'
    
    # Compile regex pattern
    regex = re.compile(pattern, flags=re.IGNORECASE)
    
    # Find all matches in the text
    matches = regex.findall(text)
    
    # Filter out matches that are "US" as a country name
    matches = [match for match in matches if match.lower() != 'us']
    
    # Count the number of personal pronouns
    count = len(matches)
    
    return count

# Folder containing input text files
input_folder = 'Aayushi/'

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        filepath = os.path.join(input_folder, filename)
        
        # Read the input text from the file
        with open(filepath, 'r', encoding='latin1') as file:
            text = file.read()
        
        # Calculate the count of personal pronouns for the current file
        pronoun_count = count_personal_pronouns(text)
        
        print(f"File: {filename}")
        print(f"Personal Pronoun Count: {pronoun_count}\n")
```

```python
import os

# Function to calculate average word length in a text
def average_word_length(text):
    # Tokenize text into words
    words = text.split()
    
    # Calculate total number of characters in all words
    total_characters = sum(len(word) for word in words)
    
    # Calculate total number of words
    total_words = len(words)
    
    # Calculate average word length
    avg_word_length = total_characters / total_words if total_words > 0 else 0
    
    return avg_word_length

# Folder containing input text files
input_folder = 'Aayushi/'

# Iterate through each file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):
        filepath = os.path.join(input_folder, filename)
        
        # Read the input text from the file
        with open(filepath, 'r', encoding='Latin1') as file:
            text = file.read()
        
        # Calculate average word length for the current file
        avg_length = average_word_length(text)
        
        print(f"File: {filename}")
        print(f"Average Word Length: {avg_length}\n")
```
